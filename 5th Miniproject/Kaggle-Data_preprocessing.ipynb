# Load Data

```python
import pandas as pd

df_train = pd.read_csv('df_train.csv')
df_train.drop('index', axis=1, inplace=True)
df_train.head()
```

# Data Preprocessing

```python
df_train[['year','month','day','weekday']]
```

```python
df_train.rename(columns={'시간대구분':'hour'}, inplace =True)
```

```python
df_train.columns
```

```python
from copy import deepcopy

df_train2 = deepcopy(df_train[['hour', '총생활인구수', 'month', 'weekday']])
```

```python
dates = []
for date in pd.date_range('2022-01-01', periods=59, freq='D'):
    dates.extend([date]*24)
```

```python
hours = []

for i in range(59):
    hours.extend([h for h in range(0, 24)])
```

```python
df_temp = pd.DataFrame({'기준일ID':dates, 'hour':hours})
```

```python
# df_temp['year'] = df_temp['기준일ID'].dt.year
df_temp['month'] = df_temp['기준일ID'].dt.month
df_temp['weekday'] = df_temp['기준일ID'].dt.dayofweek
```

```python
df_temp.head(100)
```

```python
df_total = pd.concat([df_train2, df_temp.drop('기준일ID', axis=1)], axis=0).reset_index()
df_total.drop('index', axis=1, inplace = True)
df_total
```

### Shift & Rolling

```python
df_total['2m'] = df_total['총생활인구수'].shift(1416)
df_total['2m1d'] = df_total['총생활인구수'].shift(1416+24)
df_total['2m7d'] = df_total['총생활인구수'].shift(1416+(24*7))
```

### 가변수화

```python
dumm_hour = pd.get_dummies(df_total['hour'], prefix='hour')
dumm_weekday = pd.get_dummies(df_total['weekday'], prefix='weekday')
dumm_month = pd.get_dummies(df_total['month'], prefix='month')
```

```python
df_total2 = pd.concat([df_total, dumm_hour, dumm_weekday, dumm_month], axis=1)
df_total2.drop(['hour', 'weekday', 'month'], axis=1, inplace=True)
df_total2
```

### train test split

```python
test_x = df_total2.iloc[-1416:,:].drop('총생활인구수', axis=1)
test_x
```

```python
df_total2 = df_total2.dropna()
train_x = df_total2.drop('총생활인구수', axis=1)
train_y = df_total2['총생활인구수']
```
