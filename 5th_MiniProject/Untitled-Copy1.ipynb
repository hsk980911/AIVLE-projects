```python
import pandas as pd

# 학습/평가 데이터 로딩
train_x = pd.read_csv("train_x.csv")
train_y = pd.read_csv("train_y.csv")
test_x = pd.read_csv("test_x.csv")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, StackingRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

from sklearn.metrics import *
```

```python
def test(model, filename):
    model.fit(train_x, train_y)
    
    pred = model.predict(test_x)
    result = pd.DataFrame({'id':range(0,1416), 'count':pred})
    result.to_csv(f'{filename}', index=False)
   
    print('Completed!!')
```

```python
model_cat = CatBoostRegressor()
test(model_cat, 'submission_v15.csv')
```

```python
model_xgb = XGBRegressor()

model_xgb.fit(train_x, train_y)
    
pred = model_xgb.predict(test_x)
result = pd.DataFrame({'id':range(0,1416), 'count':pred})
result.to_csv('submission_v7.csv', index=False)
```

```python
model_rf = RandomForestRegressor()
test(model_rf, 'submission_v8.csv')
```

```python
model_lgb = LGBMRegressor()
test(model_lgb, 'submission_v9.csv')
```

```python
rf = RandomForestRegressor()
xgb  = XGBRegressor()
cat = CatBoostRegressor()
cat_meta = CatBoostRegressor()

estimators = [('rf', rf), ('xgb', xgb), ('catboost', cat)]

model = StackingRegressor(estimators=estimators,
               final_estimator=cat_meta,
               verbose=1,)

model.fit(train_x, train_y).score(train_x, train_y) 
```

```python
pred = model.predict(test_x)
result = pd.DataFrame({'id':range(0,1416), 'count':pred})
result.to_csv('submission_v16.csv', index=False)
```
