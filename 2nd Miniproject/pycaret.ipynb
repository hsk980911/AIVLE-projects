```python
data = dataset.sample(frac=0.8, random_state=2022) 
data_unseen = dataset.drop(data.index) 
data.reset_index(drop=True, inplace=True) 
data_unseen.reset_index(drop=True, inplace=True) 
print('Data for Modeling: ' + str(data.shape)) 
print('Data For Predictions: ' + str(data_unseen.shape)) 
```

```python
from pycaret.classification import *
import jinja2
exp_clf = setup(data = dataset, target = 'Result_v1', session_id=2022)
```

```python
best_model = compare_models(fold = 20)
```

```python
rf = create_model('rf')
tuned_rf = tune_model(rf)

lightgbm = create_model('lightgbm')
tuned_lightgbm = tune_model(lightgbm)

xgboost = create_model('xgboost')
tuned_xgboost = tune_model(xgboost)
```

```python
blender = blend_models(estimator_list = [tuned_lightgbm, tuned_xgboost, tuned_rf])
```

```python
final_model = finalize_model(blender)
prediction = predict_model(final_model, data = data_unseen)
```
